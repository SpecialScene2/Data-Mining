#### 1-1. Review
#### 차원 축소를 하는 이유
#### - 자료의 차원은 변수의 개수로 정해지는데, 데이터마이닝 알고리즘을 효율적으로 수행하기 위해서는 변수의 수를 축소해야만 한다
#### - 일부 변수들 간 서로 높은 상관관계가 있는 변수들이나 관심 있는 결과변수와 관련 없는 변수들을 포함시킬 경우에 과적합 현상이 발생하여 모형의 정확도가 떨어질 수 있다.
#### - 변수가 많으면 정확도 하락뿐만 아니라 비용의 증가를 생각 안 할 수 없다. 따라서 변수의 수를 줄여야만 하는 것이다.

```python
# df : set_index를 통해서 'OBS#'를 index로 둔다.
df = pd.read_csv('german_classification.csv').set_index('OBS#')
```

##### Preprocessing : Dummy Variables & Scaling & Change Labels

```python
# dummy varaibles : preprocessed_df
# ['CHK_ACCT', 'HISTORY', 'SAV_ACCT', 'EMPLOYMENT', 'JOB'] : 범주형 자료들을 Dummy 변수를 만들어준다.
# PRESENT_RESIDENT는 범주형인데 NaN이라서 안쓰이는 것이다.
# DUMMY변수 만들 column들을 모아둔 후 그걸로 df을 따로 만든다
get_dummies_cols = ['CHK_ACCT', 'HISTORY', 'SAV_ACCT', 'EMPLOYMENT', 'JOB']
dummied_df = pd.get_dummies(df, columns=get_dummies_cols)

# Scaling : 수치형 자료들은 Scaling해준다.
# Scaling : min-max / min-max방법은 2진수 할 때 0,1 로 나뉘는 경우(범주형 자료)에 많이 씀 
scaling_cols = ['DURATION', 'INSTALL_RATE', 'AGE', 'NUM_CREDITS', 'AMOUNT']
preprocessed_df = dummied_df.copy()
#식 모양 : 값 - min(값) / max(값) - min(값)
preprocessed_df[scaling_cols] = (preprocessed_df[scaling_cols]-preprocessed_df[scaling_cols].min()) / (preprocessed_df[scaling_cols].max()-preprocessed_df[scaling_cols].min())

# change Label : 1,2로 응답 미응답을 표시하니 0,1로 바꿔준다.
preprocessed_df.RESPONSE = df.RESPONSE-1

# split X, y : axis=1로해야 열을 빼준다. : reponse column자체를 다 빼버리는 것임
# 신용등급 불량인 사람들(=1)이 들어있음
# split X, y 
X, y = preprocessed_df.drop('RESPONSE', axis=1), preprocessed_df.RESPONSE
```

#### Train-Test Split : `sklearn.model_selection.train_test_split`

```python
# import packages
from sklearn.model_selection import train_test_split
# split하는 이유는 새로운 앞으로의 데이터에 대해서도 답을 알맞게 구하기 위하여,
# 그 목적을 재현하기 위해 가지고 있는 데이터를 둘로 쪼개서 train은 모델 만드는 데 사용하고 
# test는 이 모델을 평가하는 데이터셋으로 사용한다.
# 여기서는 traindmf 700개에 대해서 모델 만드는 데 사용하고 앞으로의 300개 데이터에서의 
# 응답(1/0)을 정확하게 예측하는 것이 목적이다

# train_test_split with 'stratify'
# stratify = y : y의 비율인 7:3비율 파악해서 알아서 비율 맞춰서 뽑아줌(원하는 column을 써줘야함) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=23)

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

#### sklearn.linear_model.LogisiticRegression

```python
from sklearn.linear_model import LogisticRegression
# 1단계 : model을 만든다.
model = LogisticRegression()
model
```

```python
# 2단계 : parameters를 설정한다.
class_weight = 'balanced'
model = LogisticRegression(class_weight=class_weight)
#0과 1의 비율을 고려해서 만들겠다는 선언. 비율 작은 응답에 대한 샘플을 상대적으로 많이 하는 효과

#3 fit : Model에 train data들을 fit한다.
model.fit(X=X_train, y = y_train)
```

```python
from sklearn.metrics import confusion_matrix

def result(fitted_model, X_test, y_test):
    # pred 
    y_pred = fitted_model.predict(X=X_test)
    # score : 얼마나 일치하는지 알려줌.
    score = fitted_model.score(X=X_test, y=y_test)
    # conf_mat
    conf_mat = confusion_matrix(y_true=y_test, y_pred=y_pred)
    
    return [y_pred, score, conf_mat] 
# y_pred : x_test넣었을때 나오는 값, score : 얼마나 예측 잘했는지, conf_mat 을 알려준다.
```

```python
# result
# raw_result = result(raw_model, X_test, y_test)
raw_result = result(model, X_test, y_test)
raw_result
```

#### 1-3. PCA : `sklearn.decompostion.pca`

```python
# import packages
from sklearn.decomposition import pca
```

```python
# parameter : n_components
n_components = [5] #10, 15, 20] 이렇게 네 개 넣으면 네 가지 수의 변수에 대해 볼 수 있음
# 몇개에서 몇개로 줄일건지 줄일 목표를 정한다고 생각하면 됨
```

```python
print(class_weight)
```

```python
pca_model = LogisticRegression(class_weight=class_weight)
#여기서 pca_model은 단순히 변수명 아닌가?

for n in n_components:
    # decomposer : 변수갯수 줄이려는 과정
    decomposer = pca.PCA(n_components = n)
    # 30개(원래 col갯수) 중에 n개로 줄이겠다는 의미
    
    # fit
    decomposer.fit(X=X_train)
    # fit에는 X_train만 넣는다
    # Xtrain기준으로 pca돌려서 가장 좋은 n개를 뽑는다

    # .transform
    X_pca_train = decomposer.transform(X_train)
    print('X_pca_train설명', X_pca_train) #이 결과 해석: X_train은 원래 700개 row, 30개 column(변수)였음
    # 이 30개 중 5개로 변수를 줄임. decomposer가 이를 바탕으로 학습, 5차원으로 바꿔줌. 
    
    # 변수 5개로 줄여서 썼을 때 각각이 어느 정도의 분산을 설명할 수 있는지
    print('변수 5개로 줄였을때 어느정도 설명할 수 있는지?',n, decomposer.explained_variance_ratio_) 
    print(n, sum(decomposer.explained_variance_ratio_)) 
    # 42%의 분산을 설명할 수 있다. 
    # 질문! 이게 PCA1인가?
    
    # fit new estimator
    pca_model.fit(X=X_pca_train, y=y_train) #X의 차원만 바꿨다. X의 변수만.
    
    # result()
    # pca_result = result(pca_model, X_test, y_test) 
    # 위 식은 변수 30개(가변수포함 48개) 써서는 78?% 설명할 수 있다는 결과를 얻을 수 있음
    # 따라서, 
    # X_test를 그대로 넣으면 저건 변수 30개 짜리임. (가변수 포함해서 48개) 
    # X_test도 pca화 해줘야 함
    X_pca_test = decomposer.transform(X_test)
    pca_result = result(pca_model, X_pca_test, y_test)
    print(pca_result)
    #변수 30개(가변수포함 48개) 써서는 78?% 설명할 수 있고 변수 5개 써서는 68% 설명할 수 있음
```

### NLTK : Natural Languages Tool Kit
##### 1) RegexpTokenizer : Tokenizer의 종류임
- RegexpTokenizer : 정규표현식을 이용한 분리.)

##### 2) Tokenizing
- 문자열을 여러개의 조각(Token)으로 쪼개는 것.
- Token은 문자열의 한 조각으로 '단어 Token'이나 '문장 Token' 등으로 분리된다.

```python
# Tokenizing 단계
# 문자열을 여러개의 조각(Token)으로 쪼개는 것.
    
from nltk.tokenize import RegexpTokenizer

# tokenizer : RegexpTokenizer
# nltk내에서 해결해줌. tokenizer안에 들어가는 것은 특정 조건. 정규표현식. 알파벳 문자들만 잡아달라는 조건. 
tokenizer = RegexpTokenizer(r'\w+') 
# 질문! (r'\w+'가 머지???)
tokenizer.tokenize(script_rev)

tokens = [x.lower() for x in tokenizer.tokenize(script_rev)]
# tokens
tokens
```

#### Stemming이란, 간단히 말해서 단어들을 '원형'으로 표현하는 것입니다.

- 예를 들면, knowing을 know로, known을 know로 취급하는 방식을 말합니다.

※그러면 NLP에서 Stemming 작업은 왜 필요한지 생각해 봅시다. 

: 컴퓨터가 과거형 진행형들을 통해 원형을 알게하려고

사람은 knowing과 known이 결국 비슷한 의미라는걸 금방 알아차리지만 컴퓨터는 그렇지 않습니다. 

컴퓨터는 string비교를 통해서 두 string이 같은 것인지 아닌지만을 판단할 수 있기 때문입니다. 

그렇다면 컴퓨터에게 knowing과 known이 결국 같은 것이라고 인식시키는 가장 간단한 방법은 무엇일까?

답은 간단합니다. ing와 n을 빼버리는 것입니다.

```python
# stemming. 단어를 기본형으로 바꿔주는 과정(진행형, 과거형등을 원형으로 바꿔주는거)
# PorterStemmer는 단순히 stemmer의 한 종류이겠거니하고 넘어가자.
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
# .stem(token) : 질문! stem하는거겠지?
stemmed_tokens = [stemmer.stem(token) for token in tokens]
stemmed_tokens

# 원형으로 바꿧는데 written은 왜 남아있지??
```
