### 1. Data
---

#### 1-1.  `german_predictions.csv`
---
`german_predictions.csv`는 독일 은행에 신용거래를 신청한 신청자 1000명의 데이터이다. 관찰번호(OBS#)를 제외하고 총 28개의 변수로 이루어져있으며, 각 신청자들의 '신용 금액(AMOUNT)'가 예측하고자 하는 변수이다.

```python
import pandas as pd
```

```python
# 변수별 설명 : CodeList
pd.read_csv('CodeList_predictions.csv').iloc[:, :4]
```

```python
# df
df = pd.read_csv('german_predictions.csv').set_index('OBS#')
df
```

#### 1-2. Preprocessing : dummy variables

```python
# dummy varaibles : preprocessed_df
cols = ['CHK_ACCT', 'HISTORY', 'SAV_ACCT', 'EMPLOYMENT', 'JOB']
preprocessed_df = pd.get_dummies(df, columns=cols)
preprocessed_df
```

```python
# split X, y
X, y = preprocessed_df.drop('AMOUNT', axis=1), preprocessed_df.AMOUNT
```

#### 1-3. Train-Test Split : `sklearn.model_selection.train_test_split`
---
지도학습을 위해 주어진 데이터를 train set과 test set으로 나눈다.

```python
# import packages
from sklearn.model_selection import train_test_split
```

```python
# train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
X_train
```

```python
# random_state
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=23)
```

### 2. Linear Regression
---

#### 2-1. `sklearn.linear_model.LinearRegression`
---

```python
# import packages
from sklearn.linear_model import LinearRegression
```

```python
# Parameters
# fit_intercept = True # 상수를 넣을 것인지
# normalize = False # 변수들을 normalize한 상태로 모델을 얻을 것인지
```

```python
# lr_model
lr_model = LinearRegression() 
```

```python
# fit to model
lr_model.fit(X=X_train, y=y_train)
```

```python
# get predictions
y_pred_lr = lr_model.predict(X=X_test)
```

```python
# result_y_test_df 
result_y_test_df = pd.DataFrame(y_test)
result_y_test_df['lr'] = y_pred_lr
result_y_test_df
```

#### 2-2. Evaluation Measure 'RMS' : `sklearn.metrics.mean_squared_error`
---

```python
# import packages
from sklearn.metrics import mean_squared_error
import math
```

```python
# error_lr
# rms_lr
error_lr = mean_squared_error(y_pred=y_pred_lr, y_true=y_test)
rmse_lr = math.sqrt(error_lr)
rmse_lr
```

### 3. k-nn
---

#### 3-1. `sklearn.neighbors.KNeighborRegressor`
---

```python
# import packages
from sklearn.neighbors import KNeighborsRegressor
```

```python
# set parameters
n_neighbors = 5
# weights = 'uniform' # 주변 데이터의 가중치를 모두 동일하게, 'distance' 선택시, 거리에 반비례
```

```python
# knn_model
knn_model = KNeighborsRegressor(n_neighbors=n_neighbors)
```

```python
# fit to model
knn_model.fit(X=X_train, y=y_train)
```

```python
# get predictions
y_pred_knn = knn_model.predict(X=X_test)
```

```python
# result_y_test_df
result_y_test_df['knn'] = y_pred_knn
result_y_test_df
```

#### 3-2. Evaluation Measure 'RMS' : `sklearn.metrics.mean_squared_error`
---

```python
# import packages
```

```python
# error_knn
# rms_knn
error_knn = mean_squared_error(y_true=y_test, y_pred=y_pred_knn)
rmse_knn = math.sqrt(error_knn)
rmse_knn
```

#### 5-2.  K-Fold Cross Validation : `sklearn.cross_validation.cross_val_score`
---

```python
# import packages
from sklearn.cross_validation import cross_val_score
```

```python
# set parameters
estimator= lr_model
cv=5 # 전체 데이터를 k개의 그룹으로 나누고 총 k번의 모델 training - test를 반복한다.
       # 이 때 각 회차마다 test set이 되는 그룹을 계속 바꾼다.
scorer=rmse_scorer # Evaluation Measure로 사용할 Score Function - scorer(estimator, X, y)
```

```python
def rmse_scorer(estimator, X, y):
    y_pred = estimator.predict(X=X)
    rmse = math.sqrt(mean_squared_error(y_pred=y_pred, y_true=y)) 
    return rmse
```

```python
# cross_val_score
cross_val_score(estimator=estimator, cv=cv, scoring=scorer, X=X, y=y)
```
